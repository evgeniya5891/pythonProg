```python
import pandas as pd
import numpy as np
```

```python
horse_data = pd.read_csv(r'E:\PythonProg\Базовые понятия статистика\horse_data.csv', usecols=[0, 1, 3, 8, 10, 17, 19 ,22], names = ['surgery', 'Age', 'rectal_temperature', 'mucous_membranes', 'pain', 'abdomen', 'protein', 'outcome'])
horse_data.head()                     
```

```python
horse_data = horse_data.replace('?', np.nan)
horse_data.info()
```

```python
horse_data = horse_data.apply(lambda column: pd.to_numeric(column, errors='ignore'), axis=0)
```

```python
# размах – разница между минимальным и максимальным значением

rectal_temperature_range = horse_data.rectal_temperature.max() - horse_data.rectal_temperature.min()
protein_range = horse_data.protein.max() - horse_data.protein.min()
print(rectal_temperature_range)
print(protein_range)
```

Задание 2. Работа с выбросами
В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

```python
#Температура
q1 = horse_data.rectal_temperature.quantile(0.25)
q3 = horse_data.rectal_temperature.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = horse_data[horse_data.rectal_temperature.between(lower_bound, upper_bound, inclusive=True)].sort_values('rectal_temperature')
remove_outliers
#учитывая что у нас было всего 240 заполненных эл-ов, вы можем сказать что выбросов не так много, в данном случем мы должны учитывать все выбросы, т.к. мы рассматриваем больных лошадей и у них температура должна отличаться от нормы 
```

```python
#Общий белок
q1 = horse_data.protein.quantile(0.25)
q3 = horse_data.protein.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = horse_data[horse_data.protein.between(lower_bound, upper_bound, inclusive=True)].sort_values('protein')
remove_outliers
# в данном случае выпрасов нет, они все равномерно распределины по таблице , не мешает считать , так же в описании к таблице нет ограничений по белку, просто чем выше значение, тем больше обезвоживание, следовательно вероятнее смертность(предположение) 
```

```python
#rectal_temperature, я считаю что если мы заменим пропуски на среднее по таблице, результат не сильно изменится, проверим
horse_data.rectal_temperature = horse_data.rectal_temperature.replace(np.nan, 38.17)
horse_data.describe()
# Про резулитатам мы видим что некоторые показательно как мин и мах вообще не изменились, остальные изменились не значительно, что касается моды, понятно что она изменилась, но т.к. мы брали среднюю, а разброс между мин и мах не мольшой, мы можем допустит что для остальных она была такой же 
```

```python
#mucous_membranes так как это один из сиптомов, я считаю что в данном случае наши показатели можно заменить на моду, т к они свойсвены при данном заблевании 
mucous_membranes_counts = {}
for w in horse_data.mucous_membranes:
    if w not in mucous_membranes_counts:
        mucous_membranes_counts[w] = 1
    else:
        mucous_membranes_counts[w] += 1
        
        
maxw = 0
mode_mucous_membranes = None
for k, v in mucous_membranes_counts.items():
    if maxw < v:
        maxw = v
        mode_mucous_membranes = k
print('Значение моды:', mode_mucous_membranes, 'количество встречаемости:', maxw)
horse_data.mucous_membranes = horse_data.mucous_membranes.replace(np.nan, 1.0)
horse_data.describe()
```

```python
for w in horse_data.surgery:
    if w == 1:
        horse_data.pain = horse_data.pain.replace(np.nan, 4.0) 
    if w == 2:
        horse_data.pain = horse_data.pain.replace(np.nan, 2.0)
    if w == np.nan:
        horse_data.pain = horse_data.pain.replace(np.nan, 2.0)
horse_data.describe()   
# В описании бали указано что чем больше боль, том больше вероятность операции, обем от обратно и пишем 2 если операции не было и 4 если операция была 
```

```python
# в столбце живот просавим как в описании - 3-это, вероятно, препятствие, вызванное механическим ударом, по такомуже принцыпу как и боль
и обычно лечится медикаментозно
- 4 и 5 указывают на хирургическое поражение

from random import randint
for w in horse_data.surgery:
    if w == 1:
        horse_data.abdomen = horse_data.abdomen.replace(np.nan, randint(4, 5)) 
    if w == 2:
        horse_data.abdomen = horse_data.abdomen.replace(np.nan, 3.0)
    if w == np.nan:
        horse_data.abdomen = horse_data.abdomen.replace(np.nan, 2.0)
horse_data.describe()   
```
