1) Как критики относятся к спортивным играм?


H0: критики относятся хорошо к спортивным играм если средняя оценка >=7

H1: критики относятся плохо к спортивным играм если средняя оценка < 7


```python
from scipy import stats as st
alpha = 0.05
result = st.ttest_1samp(df_sport, 7)
print(result)

if (result.statistic < 0) & (result.pvalue / 2 < alpha):
    print('Отвергаем нулевую гипотезу, оценка меньше 7')
else:
    print('Не отвергаем нулевую нулевую гипотезу, критикам нравятся спортивные игры')
```

2) Критикам нравятся больше игры на PC или на PS4?

H0: критики относятся хорошо одинаково и к играм на РС и к играм на РS4

H1: критики относятся поразному к играм на РС и РS4


```python
df_PC = df.loc[(df.Platform == 'PC')]
df_PC.Critic_Score.mean()
df_PC = df_PC.Critic_Score.fillna(7.54)
```

```python
df_PS4 = df.loc[(df.Platform == 'PS4')]
df_PS4.Critic_Score.mean()
df_PS4 = df_PS4.Critic_Score.fillna(7.9)
```

```python
result = st.ttest_ind(df_PC, df_PS4, equal_var=False)
print(result)
if (result.statistic == 0) & (result.pvalue / 2 < alpha):
    print('Отвергаем нулевую гипотезу, клиенты по разному относятся к играм')
elif (result.statistic > 0) & (result.pvalue / 2 < alpha): 
    print('критикам борльше нравится игры на РС')
elif (result.statistic < 0) & (result.pvalue / 2 < alpha): 
    print('критикам борльше нравится игры на РS4')
```

3) Критикам больше нравятся стрелялки или стратегии?

H0: критики относятся хорошо одинаково и к стрелялкам и к стратегиям

```python
df_Strategy = df.loc[(df.Genre == 'Strategy')]
df_Strategy.Critic_Score.mean()
df_Strategy = df_Strategy.Critic_Score.fillna(7.43)
```

```python
result = st.ttest_ind(df_Shooter, df_Strategy, equal_var=False)
print(result)
if (result.statistic == 0) & (result.pvalue / 2 < alpha):
    print('Отвергаем нулевую гипотезу, клиенты по разному относятся к стрелялкам и стратегиям')
elif (result.statistic > 0) & (result.pvalue / 2 < alpha): 
    print('критикам борльше нравится стрелялки')
elif (result.statistic < 0) & (result.pvalue / 2 < alpha): 
    print('критикам борльше нравится стратегии')
```
