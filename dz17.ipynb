```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('2019.csv')
df.info()
```

```python

df.plot(kind='scatter', 
        x='Social_support', 
        y='GDP_per_capita', 
        title='график показывающий зависимость соц поддержкт от ввп на душу')
plt.show()
# в даном графике мы видим что соц поддержа напрямую зависит от ввп, и чем ввп больше, тем больше сох поддержка для граждан
```

```python
Country_df = df.set_index('Country_or_region')
Country_df.head()
```

```python
sng = Country_df.loc[['Armenia', 'Azerbaijan', 'Belarus', 'Georgia', 'Kazakhstan', 'Kyrgyzstan', 'Moldova', 'Russia', 'Tajikistan', 'Turkmenistan', 'Uzbekistan', 'Ukraine']].sort_values('Score')
sng.head(15)
```

```python
plt.plot(sng.index, sng.Score)
plt.title('страны и их оценка счястья жителей в странах снг')
plt.xlabel('страна')
plt.ylabel('оценка')
#на этом графике мы видим как живут страны снг, где уровень счастья жителей наименьший в Украине, а в узбекистане наивысший, несмотря на ввп и соц поддержки в этих странах
```

```python
my_top3 = Country_df[['Score']].loc[['Malawi', 'Australia', 'Russia']]
my_top3.head()
ax = my_top3.plot(kind='pie',title='круговая диограмма показывающая разность уровня счастья в странах', labels = my_top3.index, subplots=True)
```
